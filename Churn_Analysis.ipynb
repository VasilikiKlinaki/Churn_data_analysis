```python
# define a function to create a flag for high service calls

def high_service_calls_flag(x):
  if x > 3:
    return ('Yes')
  else:
    return ('No')
```

```python
dataset['high service calls'] = dataset['Customer service calls'].apply(high_service_calls_flag)
```

```python
dataset['high service calls']
```

```python
#visualizing results
dataset['high service calls'].value_counts().plot(kind='bar')
plt.show()
```

```python
#Preparing our dataset for modeling
```

```python
dataset
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
label_encoder = LabelEncoder()
```

```python
dataset['International plan'] = label_encoder.fit_transform(dataset['International plan'])
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
one_hot_encoder = OneHotEncoder()
```

```python
dataset= pd.get_dummies(dataset, columns=['State'], drop_first=True)
```

```python
bool_cols = dataset.select_dtypes(include='bool').columns
dataset[bool_cols] = dataset[bool_cols].astype(int)
```

```python
dataset
```

```python
from sklearn.model_selection import train_test_split
```

```python
x = dataset.drop(columns=['Churn','high service calls'])
y = dataset['Churn']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=42)
```

Build our first model of Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
regressor1 = LogisticRegression()
```

```python
regressor1.fit(x_train, y_train)
```

```python
y_pred = regressor1.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score, classification_report
```

```python
accuracy_score(y_test, y_pred)
```

#Build another model (random forest) to see if we can improve our results

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf_model = RandomForestClassifier(n_estimators = 80)
```

```python
rf_model.fit(x_train,y_train)
```

```python
y_pred_rf = rf_model.predict(x_test)
```

```python
accuracy_score(y_test, y_pred_rf)
```

#Let's perform hyperparameter tuning using GridSearchCV

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {
    'n_estimators': [100, 200, 300],       # number of trees
    'max_depth': [None, 5, 10, 20],        # tree depth
    'min_samples_split': [2, 5, 10],       # min samples to split node
    'min_samples_leaf': [1, 2, 4]        # min samples per leaf
}
```

```python
grid_search = GridSearchCV(
    estimator=rf_model,
    param_grid=param_grid,
    cv=5,
    n_jobs=-1,
    scoring='accuracy'
)
```

```python
grid_search.fit(x_train,y_train)
```

```python
grid_search.best_params_
```

```python
rf_model = RandomForestClassifier(max_depth = 20,
 min_samples_leaf = 1,
 min_samples_split = 2,
 n_estimators = 200)
```

```python
rf_model.fit(x_train,y_train)
```

```python
y_pred_rf = rf_model.predict(x_test)
```
